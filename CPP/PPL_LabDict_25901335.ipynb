## 2. Write a Python program to get top 3 items in a shop

```python
def top3Items(data_dict):
    ls = []
    sorted_items = sorted(data_dict.items(), key=lambda item: item[1], reverse = True)
    for i in range(3):
        ls.append(sorted_items[i])
    return ls
```

```python
sample_data = {'item1':45.5, 'item2':35, 'item3':41.3, 'item4':55, 'item5':24}
top3 = top3Items(sample_data)
```

```python
print(top3)
```

## 3. Write a Python program to find the specified number of maximum values in a given dictionary.

```python
def topNItems(data_dict, n: int):
    ls = []
    sorted_items = sorted(data_dict.items(), key=lambda item: item[1], reverse = True)
    for i in range(n):
        ls.append(sorted_items[i])
    return ls
```

```python
sample_data = {'item1':45.5, 'item2':35, 'item3':41.3, 'item4':55, 'item5':24}
top4 = topNItems(sample_data, 4)
```

```python
print(top4)
```

## 4. WAP to combine two lists into a dictionary. The elements in the first one serve as keys and the elements of the second one serve as values. Each item in the first list must be unique and hashable.

```python
def combineListsToDic(keys, vals):
    setOfKeys = set(keys)
    if(len(setOfKeys)!=len(keys)):
        print("The first list isn't hashable.")
        return None
    
    if(len(keys)!=len(vals)):
        print('WARNING: The lists of keys and values arent of the same length.')
        
    return dict(zip(keys, vals))
```

```python
ls1 = [1,2,3,4,5]
ls2 = ['a','b','c','d','e']
dictTest = combineListsToDic(ls1,ls2)
```

```python
print(dictTest)
```

## 5. Write a Python function that evaluates math expressions represented as dictionaries. The expression dictionary should support basic operations (+,-,*,/), variables, and nested expressions

```python
def evaluate_expr(expr, variables):
    if isinstance(expr, (int, float)):
        return expr
    
    if isinstance(expr, str):
        return variables[expr]
    
    op = expr["op"]
    left = evaluate_expr(expr["left"], variables)
    right = evaluate_expr(expr["right"], variables)

    if op == "+":
        return left + right
    elif op == "-":
        return left - right
    elif op == "*":
        return left * right
    elif op == "/":
        return left / right
    else:
        raise ValueError("Unknown operator: " + op)
```

```python
expr = {
    "op": "+",
    "left": 5,
    "right": {
        "op": "*",
        "left": "x",
        "right": 3
    }
}
```

```python
variables = {"x": 4}
```

```python
result = evaluate_expr(expr, variables)
print(result)
```
