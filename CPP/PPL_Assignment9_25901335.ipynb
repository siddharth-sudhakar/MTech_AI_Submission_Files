## Question 4: Simulate and analyze traffic movement on a single road with several lanes using lists.

- Generate random vehicles
- Simulate entry and exit
- Measure congestion, flow, and average wait time

(Show output as a report)

```python
import random
```

```python
num_lanes = 3
lane_length = 10          # positions per lane
timesteps = 20            # how long to simulate
entry_chance = 0.5        # chance a new vehicle enters per step
```

```python
road = [[] for _ in range(num_lanes)]   # list of lanes, each lane = list of vehicles
all_vehicles = []                       # keep record of all vehicles
exited_vehicles = []                    # store vehicles that exited
```

```python
vehicle_id = 1
```

```python
for t in range(timesteps):
    # Move vehicles forward and handle exits
    for lane in road:
        for v in lane[:]:
            v["position"] += v["speed"]
            if v["position"] >= lane_length:  # exit condition
                v["exit_time"] = t
                exited_vehicles.append(v)
                lane.remove(v)
    
    # Randomly add new vehicles
    if random.random() < entry_chance:
        lane_idx = random.randint(0, num_lanes - 1)
        new_vehicle = {
            "id": vehicle_id,
            "lane": lane_idx,
            "position": 0,
            "speed": random.randint(1, 3),
            "entry_time": t
        }
        road[lane_idx].append(new_vehicle)
        all_vehicles.append(new_vehicle)
        vehicle_id += 1
    
    # Print snapshot of traffic
    print(f"Time {t}: {[len(l) for l in road]} vehicles per lane")
```

```python
total_exited = len(exited_vehicles)
total_generated = len(all_vehicles)
avg_congestion = sum(len(lane) for lane in road) / num_lanes
avg_flow = total_exited / timesteps
```

```python
if total_exited > 0:
    wait_times = [v["exit_time"] - v["entry_time"] for v in exited_vehicles]
    avg_wait = sum(wait_times) / total_exited
else:
    avg_wait = 0
```

```python
print("\n=== Traffic Simulation Report ===")
print(f"Total vehicles generated : {total_generated}")
print(f"Total vehicles exited    : {total_exited}")
print(f"Average flow rate        : {avg_flow:.2f} vehicles/time step")
print(f"Average congestion       : {avg_congestion:.2f} vehicles/lane")
print(f"Average wait time        : {avg_wait:.2f} time steps")
```
