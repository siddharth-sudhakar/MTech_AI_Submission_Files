Q1. Write a program to display data of different types using variables and literal constants. 

Objective:
To store values in variables of different datatypes and print them.

```python
name = "Rohan"       # string
age = 20             # integer
height = 5.9         # float
is_student = True    # boolean

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)
```

Explanation:
Variables are created for string, integer, float and boolean. print() displays them on screen.

Q2. Write a program to show the application of the += operator on strings

Objective:
To show how += can extend a string.

```python
msg = "Hello"
msg += " World"     # adding using +=

print(msg)
```

Explanation:
A string variable is made and extra text is attached to it using += operator.

Q3. Write a program that performs addition and multiplication on string variables.

Objective:
To understand how strings behave with + and * operators.

```python
name = "Hi"
print(name + "!!!")     # string addition
print(name * 3)         # string repetition
```

Explanation:
'+' joins two strings and '*' repeats a string multiple times.
