Q1. Write a Python program that uses a lambda function to:
- Take a number as input.
- Use a lambda function to square the number.
- Print the squared value.

Objective:
To use a lambda function for a small calculation.

```python
n = int(input("Enter number: "))

sq = lambda x: x*x
print("Square:", sq(n))
```

Explanation:
Lambda expression returns square of the input number.

Q2. Write a recursive function count digits(n) that returns the number of digits in a positive integer n.

Objective:
To count digits using recursion.

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n//10)

x = int(input("Enter number: "))
print("Digits =", count_digits(x))
```

Explanation:
Function divides number by 10 repeatedly until zero.

Q3. Write a Python program to:
- Take n integers into a list.
- Print the list in original order.
- Print the list in reverse order (without using [::-1] or reverse()).

Objective:
To accept numbers in a list and print them in both orders.

```python
nums = []

n = int(input("How many numbers: "))

for i in range(n):
    nums.append(int(input()))

print("Original:", nums)

print("Reverse:", end=" ")
for i in range(len(nums)-1, -1, -1):
    print(nums[i], end=" ")
```

Explanation:
Normal loop prints list, reverse loop prints from last to first.
