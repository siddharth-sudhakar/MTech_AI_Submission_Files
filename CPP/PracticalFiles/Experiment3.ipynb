Q1. Write a Python program that asks the user to enter an integer in decimal (base 10) and then prints its equivalent in binary, octal, and hexadecimal using built-in functions.

Objective:
To convert decimal number to other number systems.

```python
num = int(input("Enter a decimal number: "))

print("Binary :", bin(num))
print("Octal  :", oct(num))
print("Hex    :", hex(num))
```

Explanation:
Built-in bin(), oct() and hex() show binary, octal and hexadecimal equivalent.

Q2. Write a Python program that reads a string representing a floating-point number (for example, "25.67") from the user, converts it to a float, adds 10.5 to it, and prints the result.

Objective:
To convert a string into float and perform arithmetic.

```python
v = input("Enter floating number: ")

x = float(v)
x = x + 10.5

print("Result =", x)
```

Explanation:
String input converted using float() and then 10.5 is added.
