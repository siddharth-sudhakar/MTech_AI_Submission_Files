Q1. Write a program to calculate area of a triangle using Heron’s Formula.

Objective:
To calculate triangle area using 3 sides.

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2        # semi-perimeter
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

print("Area:", area)
```

Explanation:
Semi-perimeter is found first and then Heron’s formula is applied using square root.

Q2. Write a program to calculate the total amount of money in the piggy bank, given
the coins of Rs 10,5,2,1.

Objective:
To find total money using coin values.

```python
r10 = int(input("10 rs coins: "))
r5  = int(input("5 rs coins: "))
r2  = int(input("2 rs coins: "))
r1  = int(input("1 rs coins: "))

total = r10*10 + r5*5 + r2*2 + r1*1
print("Total:", total)
```

Explanation:
User gives number of coins, each value multiplied and then added.
