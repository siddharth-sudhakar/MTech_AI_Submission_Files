Q1. Write a program to print the following pattern:

1

12

123

1234

12345

Objective:
To print a numeric pattern using nested loops.

```python
for i in range(1, 6):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

Explanation:
Outer loop controls rows, inner loop prints numbers in each row.

Q2. Write the program to print multiplication table of n, where n is entered by the user

Objective:
To revise loop logic using multiplication table.

```python
n = int(input("Enter n: "))
print()
for i in range(1, 11):
    print(n, "*", i, "=", n*i)
```

Explanation:
Simple for loop prints n Ã— i format.

Q3. Write a Python function student info that takes three keyword arguments: name, age, and course. Call the function using keyword arguments in any order and print the information in a readable format.

Objective:
To learn function calls using keyword parameters.

```python
def student_info(name, age, course):
    print("Name :", name)
    print("Age  :", age)
    print("Course:", course)
```

```python
student_info(age=20, course="BCA", name="Rohan")
```

Explanation:
Function parameters are passed in random order using key names.

Q4. Write a program that prints absolute value, square root and cube of a number.

Objective:
To compute absolute, root and cube.

```python
import math

x = float(input("Enter number: "))

print("Absolute:", abs(x))
print("Square root:", math.sqrt(x))
print("Cube:", x*x*x)
```

Explanation:
abs(), math.sqrt() and multiplication are used to show calculations.
