## String and its operation

Q1. Write a program that takes a string as input and prints a new string where all vowels are removed, but the order of the remaining characters is preserved.

Objective:
To filter out vowels from a string.

```python
text = input("Enter text: ")
out = ""

for ch in text:
    if ch.lower() not in "aeiou":
        out += ch

print(out)
```

Explanation:
Characters checked one by one and only consonants stored.

Q2. Given a string, write a program to check whether it is a palindrome or not, ignoring case and all non-alphanumeric characters.

Objective:
To check if a given text reads same backwards.

```python
s = input("Enter string: ")

clean = ""
for c in s:
    if c.isalnum():
        clean += c.lower()

if clean == clean[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

Explanation:
String cleaned and compared to its reverse form.

Q3. Write a program that takes a sentence as input and prints the words in reverse order, while keeping the characters inside each word in the same order.

Objective:
To reverse words in a sentence.

```python
line = input("Enter sentence: ")
words = line.split()
words.reverse()

for w in words:
    print(w, end=" ")
```

Explanation:
split() creates list, reversed list printed word by word.
