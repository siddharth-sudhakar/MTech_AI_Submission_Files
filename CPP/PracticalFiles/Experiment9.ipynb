## File Handling

Q1. Write a program that counts the number of tabs, spaces and newline characters in a file.

Objective:
To count spaces, tabs and newlines in a file.

```python
fname = "data.txt"

tab = space = newline = 0

with open(fname, "r") as f:
    for ch in f.read():
        if ch == "\t":
            tab += 1
        elif ch == " ":
            space += 1
        elif ch == "\n":
            newline += 1

print(tab, space, newline)
```

Explanation:
Characters read in loop and counters updated using conditions.

Q2. Write a program that copies first 10 bytes of a binary file into another.

Objective:
To copy first few bytes of one binary file to another.

```python
with open("pic.jpg", "rb") as f1:
    data = f1.read(10)

with open("copy.bin", "wb") as f2:
    f2.write(data)
```

Explanation:
read() reads limited bytes and write() stores them in new file.
