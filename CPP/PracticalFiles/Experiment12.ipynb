## Operator Overloading, Error Handling

Q1. Write a program that overloads the + operator on a class Student that has
attributes name and marks.

Objective:
To overload + operator for Student objects.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks

s1 = Student("A", 90)
s2 = Student("B", 80)

print(s1 + s2)
```

Explanation:
__add__() returns combined marks of two student objects.

Q3. Write a program that asks the user to enter an integer. If the user enters something that is not an integer (like "abc" or 3.5), the program should catch the error, print "Invalid input, please enter an integer.", and ask again until a valid integer is entered.

Objective:
To validate user input using exception handling.

```python
while True:
    try:
        x = int(input("Enter integer: "))
        break
    except:
        print("Invalid input, try again")

print("You entered:", x)
```

Explanation:
try–except keeps asking user until valid integer is entered.
