Q3. Write a program to print the reverse of a number.

Objective:
To reverse digits using a loop.

```python
n = int(input("Enter number: "))

rev = 0
temp = n

while temp > 0:
    rev = rev*10 + temp%10
    temp //= 10

print("Reverse:", rev)
```

Explanation:
Digits are extracted using modulus and placed in reverse order.

Q4. Write the program to print multiplication table of n, where n is entered by the user.

Objective:
To print table of a number using loop.

```python
n = int(input("Enter n: "))
print()
for i in range(1, 11):
    print(n, "x", i, "=", n*i)
```

Explanation:
for loop prints values from 1 to 10 multiplied by the given number.
