Q1. Write a Python program to:
i. Take n integers from the user and store them in a list.
ii. Create a new list that contains the same elements but without duplicates.
iii. Print both the original list and the new list.

Objective:
To remove repeated values from list.

```python
n = int(input("Enter count: "))
nums = []

for i in range(n):
    nums.append(int(input()))

new = list(set(nums))

print("Original:", nums)
print("Unique  :", new)
```

Explanation:
Original list converted to a set to eliminate duplicates.

Q2. Write a Python program to:
i. Create a tuple, e.g. nums = (10, 20, 30, 40, 50).
ii. Ask the user to enter a number.
iii. Print the index of that number in the tuple if it exists, otherwise print a message saying it is not present.

Objective:
To search for an item inside a tuple.

```python
nums = (10,20,30,40,50)

x = int(input("Enter number: "))

if x in nums:
    print("Index:", nums.index(x))
else:
    print("Not present")
```

Explanation:
in operator checks presence; if true index is printed.

Q3. Write a program that reads two sequences of integers (allow duplicates)
and prints:
i. the set of elements common to both sequences,
ii. the set of elements that appear in exactly one of the two sequences (symmetric difference). Elements in each output set should be printed in sorted order without duplicates.

Objective:
To find intersection and symmetric difference between sequences.

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))

A = set(a)
B = set(b)

print(sorted(A & B))
print(sorted(A ^ B))
```

Explanation:
Lists converted to sets and operations & and ^ applied.
