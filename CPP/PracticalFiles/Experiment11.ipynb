Q1. Write a program with class Employee that keeps a track of the number of employees in an organization and also stores their name, designation and salary details.

Objective:
To store employee details using objects and count total employees.

```python
class Employee:
    count = 0

    def __init__(self, name, des, sal):
        Employee.count += 1
        self.name = name
        self.des = des
        self.sal = sal

e1 = Employee("Ram","Manager",50000)
e2 = Employee("Raj","Clerk",25000)

print("Total employees:", Employee.count)
```

Explanation:
Class variable increments inside constructor whenever new object created.

Q2. Write a program that has classes such as Student, Course, and Department.
- Enroll a student in a course of a particular department.
- Student Class (name, rollno, course, year)
- Course Class (name, year)
- Department Class (name)

Objective:
To use separate classes to represent academic relationship.

```python
class Department:
    def __init__(self, name):
        self.name = name

class Course:
    def __init__(self, name, year):
        self.name = name
        self.year = year

class Student:
    def __init__(self, name, roll, course, year):
        self.name = name
        self.roll = roll
        self.course = course
        self.year = year

d = Department("Computer Science")
c = Course("Python", 2025)
s = Student("Rohan", 12, c.name, c.year)

print(s.name, s.roll, s.course, s.year)
```

Explanation:
Student object stores course and department related information.
