## **Question 2:**
Write a program that simulates mutation on an immutable string without using:
- replace()
- list()
- join()

Replace every k-th occurence of a given character with another character.

**Objective:** Given a string, replace every k-th occurence of a given character with another character.

```python
def mutateStr(string, oldChar, newChar, k):
    count = 0
    i = 0
    while i < len(string): # Loop through the string
        if string[i] == oldChar: # Identifies the character to be replaced in the string
            count += 1 # Keeps track of count for the specified character
            if count % k == 0: # If the count hits the k-th occurence
                string = string[:i] + newChar + string[i+1:] # Replace the k-th occurence with the other specified character
        i += 1
    return string
```

```python
mutateStr("Sidddharth", 'd', 'f', 2)
```

**Explanation:** This mutation algorithm uses string concatenation concept. Here the k-th occurence is replaced by reassigning the updated string to the original string in the required format using string slicing and concatenation.

## **Question 3:**
Write a program that prints a number, its square and its cube repeatedly in the range (1,n)

**Objective:** To write a program that given a number n, it prints a number, its square and its cube repeatedly in range (1,n)

```python
def printNumSqCube(n):
    for i in range(1,n+1): # Loop to traverse in the range (1,n)
        print("The number is:", i) # Print the number i
        print("The number to the power of 2 is:", i**2) # Print the number i's square value
        print("The number to the power of 3 is:", i**3) # Print the number i's cube value
        print()
```

```python
printNumSqCube(4)
```
