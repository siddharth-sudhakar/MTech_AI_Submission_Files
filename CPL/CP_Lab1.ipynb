## 3. "Hello Kernel" CUDA Program

```python
@cuda.jit
def hello_kernel():
  # Thread and block indices
  tx = cuda.threadIdx.x
  bx = cuda.blockIdx.x
  bdim = cuda.blockDim.x

  # Global thread ID
  gid = bx * bdim + tx

  print("Hello from Block", bx, "Thread", tx, "Global ID", gid)
```

```python
blocks = 2
threads_per_block = 4
```

```python
hello_kernel[blocks,threads_per_block]()
cuda.synchronize()
```
